Seaborn 

Scatter plot = scatter plot can be plotted for both numerical as well as  categorical data, it's mostly used the relationship between two data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head()
```

            Plot a scatter plot between total_bil and  tip

```python
# plot here

sns.relplot(x = 'total_bill' , y = 'tip', kind = 'scatter' , data = data)
```

        The hue parameter

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = data)
```

                     The style parameter

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker',style = 'sex', data = data)
```

        The size parameter 

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'sex', size = 'size', data = data )
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'sex', size = 'size', sizes = (15,200), data = data )
```

        The scatter plot function

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'sex', data = data)
```

             Categorical Scatterplots

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.style.use('fivethirtyeight')

sns.catplot(x = 'day', y = 'tip', kind = 'strip',hue = 'sex', data = tips, palette="Set2")
#jitter
```

jitter

```python
sns.catplot(x = 'day', y = 'tip', kind = 'strip', jitter = 0, data = tips, palette="Set2")
```

```python
sns.catplot(x='day', y='tip', kind='swarm', data=tips, palette="Set2")
```

```python
sns.catplot(x = 'day', y = 'tip', hue = 'sex', kind  = 'swarm', data = tips)
```

```python
sns.swarmplot(x = 'day', y = 'tip', hue = 'sex', data = tips)
```

BOX PLOT

```python
#UNIVIRIATE
sns.boxplot(x =tips['total_bill'])
```

```python
plt.xticks(rotation=90)
sns.boxplot(x= tips['tip'])
plt.show()
```

```python
sns.catplot(x= 'day', y = 'total_bill', kind = 'box', data = tips)
```

```python
sns.boxplot(x= 'day', y = 'total_bill', data = tips)
```

```python
sns.boxplot(x = tips["day"], y = tips['total_bill'])
```

```python
sns.catplot(x= 'day', y = 'total_bill',hue = 'smoker', kind = 'box', data = tips)
```

violin Plot

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x = 'day', y = 'total_bill',kind = 'violin' , data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill' , hue = 'sex', kind = 'violin',  split = False, data = tips )
```

```python
sns.catplot(x = 'day', y = 'total_bill' , hue = 'sex', kind = 'violin',  split = True, data = tips )
```

Barplot and countplot

```python
sns.catplot(x = 'smoker', y = 'total_bill', kind = 'bar', data = tips)
```

```python
sns.catplot(x = 'smoker', y = 'total_bill', hue = 'sex', kind = 'bar', data = tips)
```

```python
sns.barplot(x = 'smoker', y = 'total_bill', hue = 'sex', estimator=np.std, data = tips)
```

```python
sns.catplot(x = 'sex' , y = 'total_bill', hue = 'smoker', kind = 'bar', data = tips)
```

```python
sns.countplot(x = 'sex', hue = 'smoker', data = tips)
```

COUNT PLOT

```python
sns.catplot(x = 'sex' , hue = 'smoker', kind = "count", data = tips)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

JOINTPLOT

```python
sns.jointplot(x = 'total_bill', y = 'tip', kind = 'hex', data = tips)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', kind = 'kde', data = tips)
```
